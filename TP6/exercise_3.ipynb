**Etape 1:** Chargement des donn√©es

```python
import pandas as pd

scores_df = pd.read_csv('eurovision-2016-televoting.csv', index_col=0)
country_names = list(scores_df.index)
scores_df
```

**Etape 2:**



Affichez le DataFrame et Analyser. Chaque ligne repr√©sente un pays qui a vot√©.

Remarquez les valeurs NaN ("not-a-number"). Ceux-ci correspondent aux scores manquants dans le fichier CSV d'origine. Ces scores sont manquants car les pays ayant vot√© pour eux m√™mes alors qu'ils n'ont pas √©t√© autoris√©s √† le faire.

```python
scores_df
```

**Step 3:** Remplacer les NaN avec le score le plus √©lev√© possible (12) - nous supposons que les pays voteraient pour eux-m√™mes, s'ils y √©taient autoris√©s.

```python
df_no_na= scores_df.fillna(12)
```

**Step 4:** Importer les fonctions  `normalize` from `sklearn.preprocessing`.

```python
 from sklearn.preprocessing import Normalizer
```

**Etape 5:** 


Appliquez la fonction de normalisation √† `scores_df.values` en affectant le r√©sultat √† `samples`.

(Pourquoi devons-nous normaliser? Parce que maintenant que les valeurs manquantes sont remplies avec la valeur de de 12 (points), on risque de d√©passer les intervalles initialement d√©finis. 

```python
transformer = Normalizer().fit(df_no_na)
transformer
X = transformer.transform(df_no_na)
```

**Step 6:** Import des classes necessaires:
 + `linkage` and `dendrogram` from `scipy.cluster.hierarchy`.
 + `matplotlib.pyplot` as `plt`.

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

**Etape 7:** Effectuez un clustering hi√©rarchique sur `samples` en utilisant la fonction` linkage () `avec l'argument` method = 'single'`. 

```python
Z = linkage(X, 'single')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z)
plt.show()
```

**Step 8:** Plot a dendrogram of the hierarchical clustering, using the list `country_names` as the `labels`. In addition, specify the `leaf_rotation=90`, and `leaf_font_size=6` keyword arguments as you have done earlier.


Tracez un dendrogramme du clustering hi√©rarchique, en utilisant `country_names` comme `labels`. 
De plus, sp√©cifiez les arguments  `leaf_rotation=90`, et `leaf_font_size=6`.

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Calculate the linkage: mergings
mergings = linkage(X, method='single')

# Plot the dendrogram
dendrogram(mergings, labels=country_names, leaf_rotation=90, leaf_font_size=6)
plt.show()
```
